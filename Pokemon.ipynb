## Playing around data

```python
# Fastest pokemon overall
pokemon["Fastest"]= (pokemon["Speed"]/pokemon["Speed"].mean()).sort_values()
```

```python
pokemon.sort_values(by="Fastest", ascending=False).reset_index(drop=True)
```

```python
pokemon.drop(columns=["Fastest"],inplace=True)
```

```python
#Total power of pokemon(Hp + Attackt + Defense + SP.Atk/Def + Speed)

pokemon["Total"]= pokemon.iloc[:,4:10].sum(axis=1)
pokemon
```

#Filtering

```python
#Total power comparasion whether pokemon is Legand or not 
pokemon.groupby("Legendary")["Total"].agg(sum)
```

```python
#Pokemons names which include "Mega"
pokemon.loc[pokemon["Name"].str.contains("Mega")]
```

```python
# All type 1 of pokemon except include "Fire"
pokemon.loc[~pokemon["Type 1"].str.contains("Fire")]
```

```python
pokemon.describe()
```

```python
# Assign pokemon as "quite legendary" if attack and defense higher than mean+std
## Assign pokemon as "legendary" if attack and defense higher than mean


pokemon.loc[(pokemon["Attack"] > 80) & (pokemon["Defense"] > 73), ("Legendary")] = "Legend"
pokemon.loc[(pokemon["Attack"] > 121) & (pokemon["Defense"] > 104), ("Legendary")] = "Quite Legend"
pokemon.loc[(pokemon["Attack"] <= 80) | (pokemon["Defense"] <= 73), ("Legendary")] = "Poor Babe"


```

```python
#Number of legendary conditions 
pokemon.groupby("Legendary")["Total"].count()
```

```python
pokemon["Super Attack"] = pokemon["Attack"]+pokemon["Sp. Atk"]
pokemon["Super Defense"] = pokemon["Defense"]+pokemon["Sp. Def"]
```

```python
df = pokemon.sort_values(by=["Super Attack", "Super Defense"],ascending=False).reset_index(drop=True)
```
